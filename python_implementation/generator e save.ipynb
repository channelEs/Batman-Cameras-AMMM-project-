```python
import random
import sys
import os

def generate_instance(N, K, filename=None):
    """
    Genera un'istanza del problema Batman in formato OPL .dat
    e la salva direttamente su un file.
    """
    
    # Se non specificato, crea un nome file automatico es: batman_N10_K3.dat
    if filename is None:
        filename = f"batman_N{N}_K{K}.dat"
    
    print(f"\nGenerazione istanza con N={N}, K={K}...")
    
    # --- 1. Generazione Dati ---
    P = [random.randint(10, 100) for _ in range(K)]
    R = [random.randint(1, 49) for _ in range(K)]
    A = [random.randint(2, 6) for _ in range(K)]
    C = [round(random.uniform(1.0, 10.0), 2) for _ in range(K)]

    M = [[0 for _ in range(N)] for _ in range(N)]
    for i in range(N):
        for j in range(i + 1, N):
            is_reachable = random.random() > 0.3
            if is_reachable:
                dist = random.randint(1, 49)
            else:
                dist = random.randint(50, 100)
            M[i][j] = dist
            M[j][i] = dist

    # --- 2. Scrittura su File ---
    try:
        with open(filename, 'w') as f:
            f.write("/*************************************************\n")
            f.write(f" * Istanza Generata Random\n")
            f.write(f" * N (Incroci) = {N}\n")
            f.write(f" * K (Modelli) = {K}\n")
            f.write(" *************************************************/\n\n")
            
            f.write(f"N = {N};\n")
            f.write(f"K = {K};\n\n")
            
            f.write(f"P = {P};\n")
            f.write(f"R = {R};\n")
            f.write(f"A = {A};\n")
            f.write(f"C = {C};\n\n")
            
            f.write("M = [\n")
            for row in M:
                f.write(f"  {row},\n")
            f.write("];\n")
            
        print(f"[SUCCESSO] File creato correttamente: {filename}")
        print(f"Trovi il file nella stessa cartella di questo script.")
        
    except Exception as e:
        print(f"\n[ERRORE] Impossibile scrivere il file: {e}")

if __name__ == "__main__":
    # Se lanciato senza argomenti (es. da IDE o doppio click), chiede input all'utente
    if len(sys.argv) != 3:
        print("--- Generatore Istanze Batman (OPL .dat) ---")
        try:
            val_n = input("Inserisci il numero di Incroci (N): ")
            val_k = input("Inserisci il numero di Modelli (K): ")
            
            n_input = int(val_n)
            k_input = int(val_k)
            
            generate_instance(n_input, k_input)
        except ValueError:
            print("Errore: Devi inserire numeri interi validi.")
        
        # Pausa finale per non far chiudere la finestra subito su Windows
        input("\nPremi INVIO per chiudere...")
        
    else:
        # Modalità terminale (se volessi ancora usarla): python generator.py 10 3
        try:
            n_input = int(sys.argv[1])
            k_input = int(sys.argv[2])
            generate_instance(n_input, k_input)
        except ValueError:
            print("// Errore: N e K devono essere numeri interi.")
```

```python
import csv
import os
import sys

# Nome del file dove salvare i risultati
FILE_NAME = "Risultati_Progetto_Batman.csv"

def init_file():
    """Crea il file con l'intestazione se non esiste"""
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, mode='w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f, delimiter=';')
            # Intestazioni delle colonne (Header)
            writer.writerow(["N (Incroci)", "K (Modelli)", "Obiettivo (Costo)", "Tempo (sec)"])
        print(f"[INFO] Creato nuovo file: {FILE_NAME}")

def main():
    init_file()
    
    print(f"\n--- Logger Manuale Risultati CPLEX ---")
    print(f"I dati verranno salvati in: {FILE_NAME}")
    print("Istruzioni: Esegui CPLEX, copia la riga che inizia con 'DATA:', incollala qui e premi Invio.")
    print("Scrivi 'exit' per chiudere il programma.\n")

    while True:
        try:
            user_input = input("Incolla riga CPLEX > ").strip()
            
            if user_input.lower() in ['exit', 'esci', 'q']:
                print("Chiusura programma. Buon lavoro!")
                break
            
            if not user_input:
                continue

            # Pulizia della stringa (toglie "DATA:" se presente)
            clean_input = user_input.replace("DATA:", "").strip()
            
            # Parsing dei dati
            parts = clean_input.split(';')
            
            if len(parts) != 4:
                print("⚠️  ERRORE: Formato non valido.")
                print("   Assicurati di copiare la riga intera tipo: DATA:10;3;321.78;0.05")
                continue
                
            # Estrazione variabili per feedback visivo
            n, k, costo, tempo = parts
            
            # Salvataggio nel file (Append mode 'a')
            with open(FILE_NAME, mode='a', newline='', encoding='utf-8') as f:
                writer = csv.writer(f, delimiter=';')
                writer.writerow(parts)
                
            print(f"✅ Salvato! [N={n}, K={k}, Costo={costo}] -> Aggiunto al file.")

        except Exception as e:
            print(f"❌ Errore imprevisto: {e}")
            
if __name__ == "__main__":
    main()
```
